```python
# X = infinite
X = float('inf')

w = [
    #A  B  C  D  E  F  G  H
    [X, 2, X, X, X, 1, X, X],  # A
    [2, X, 2, 2, 4, X, X, X],  # B
    [X, 2, X, X, 3, X, X, 1],  # C
    [X, 2, X, X, 3, 1, X, X],  # D
    [X, 4, 3, 3, X, X, 7, X],  # E
    [1, X, X, 1, X, X, 5, X],  # F
    [X, X, X, X, 7, 5, X, 6],  # G
    [X, X, 1, X, X, X, 6, X]   # H
]
w2 = [
    [X, 3, 4, 5, 6],
    [3, X, 6, 5, 6],
    [4, 6, X, 3, 4],
    [5, 5, 3, X, 2],
    [6, 6, 4, 2, X]
]
w3 = [
    [X, 2, X, X, X, 1, X, X],
    [2, X, 2, 2, 4, X, X, X],
    [X, 2, X, X, 3, X, X, 1],
    [X, 2, X, X, 4, 1, X, X],
    [X, 4, 3, 4, X, X, 7, X],
    [1, X, X, 1, X, X, 5, X],
    [X, X, X, X, 7, 5, X, 6],
    [X, X, 1, X, X, X, 6, X]
]
w4 = [
    [X, 2, X, X, X, 8, X, X],
    [2, X, 2, 2, 4, X, X, X],
    [X, 2, X, X, 3, X, X, 1],
    [X, 2, X, X, 3, 1, X, X],
    [X, 4, 3, 3, X, X, 7, X],
    [8, X, X, 1, X, X, 5, X],
    [X, X, X, X, 7, 5, X, 6],
    [X, X, 1, X, X, X, 6, X]
]
w5 = [
    [X, 1, 9, X],
    [1, X, X, 9],
    [9, X, X, 1],
    [X, 9, 1, X]
]
w6 = [
    [X, 5, X, 3],
    [5, X, 2, 1],
    [X, 2, X, X],
    [3, 1, X, X]
]

# returns minimal cost by spanning to all nodes only once
def prim(w):
    # skapa variabel med kostnaden 0
    cost = 0
    # skapa en Set för alla noder som blir anslutna
    connected = set()
    # lägg in första noden som ansluten
    connected.add(0)

    # kör så länge alla noder inte är anslutna ännu
    while len(connected) < len(w):
        # variabel med max kostnad hittils
        min_cost = X
        min_node = -1
        # kör igenom alla noder
        for i in range(len(w)):
            # om noden inte är ansluten
            if i not in connected:
                # gå igenom alla anslutningar
                for j in range(len(w)):
                    # om anslutningen inte är ansluten 
                    # och kostnaden är lägre än den hittils lägsta
                    # sätt den till noden med lägst kostnad
                    if j in connected and w[i][j] < min_cost:
                        min_cost = w[i][j]
                        min_node = i

        # addera den minsta hittade kostnaden
        cost += min_cost
        # markera noden som ansluten
        connected.add(min_node)
    # returnera kostnaden till slut
    return cost

print(prim(w))
print(prim(w2))
print(prim(w3))
print(prim(w4))
print(prim(w5))
print(prim(w6))
```
